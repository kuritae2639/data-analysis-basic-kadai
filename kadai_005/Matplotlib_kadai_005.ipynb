```python
import numpy as np
import matplotlib.pyplot as plt

# 描画領域に6領域(2行3列)のサブプロットを作成する
fig, axes = plt.subplots(2, 3, tight_layout=True)

# 折れ線グラフ plt.plot(y)
x1 = [0,1,2,3,4]
y1 = [2,5,3,7,2]
axes[0,0].plot(y1)

# sin関数
x2 = np.linspace(0,10,150)
y2 = 3 + np.sin(2 * x2)

axes[0,1].plot(x2,y2,linewidth=2.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 11),
              ylim=(0, 5), yticks=np.arange(0, 6))

# ヒストグラム plt.hist()
x3 = np.random.normal(20,3,100)

axes[0,2].hist(x3,bins=30)

# 散布図 plt.scatter()
np.random.seed(4)
x4 = 5 + np.random.normal(0,2,50)
y4 = 5 + np.random.normal(0,2,len(x4))

axes[1,0].scatter(x4,y4)

plt.show()
```
